Q6. Write a code to perform basic string compression using the counts of a repeated characters.

```python
s1= "aaabbbbccccdddefff"
compressed_string= ""
count=1
for i in range(1, len(s1)):
  if s1[i] == s1[i-1]:
    count+=1
  else:
    compressed_string += s1[i-1] + str(count)
    count=1
compressed_string += s1[-1] + str(count)
compressed_string
```

Q20. Write a code to shuffle a given list randomly without using any built-in shuffle functions.

```python
import random

lis = [40,30,130,50,39,20,50,604,30]
for i in range(len(lis)-1,0,-1):
    index = random.randint(0,i)
    lis[i],lis[index] = lis[index],lis[i]
print("shuffled list = ",lis)
```

Q31. Write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list.

```python
words = ["tomato","potato","tomato","potato","potato","onion","carrot","carrot","raddish","raddish"]
count=1
word = sorted(words)
dictionary={}
for i in range(1,len(words)):
    if word[i] == word[i-1]:
        count+=1
    else:
        dictionary[word[i-1]]=count
        count=1
    dictionary[word[i]]=count
print(dictionary)
```

  Q32. Write a code that takes two dictionaries as an input and merges them into a single dictionary. If there are common keys, the value should be added together.

```python
dict1 = {"name":"rashi","study":"data science","interests":"singing"}
dict2 = {"age":20, "interests":"travelling"}
for key1,value1 in dict1.items():
  for key2,value2 in dict2.items():
    if key1==key2:
      dict2[key1]= (value1,value2)
final_dictionary = dict1 | dict2
print("the final dict is : ",final_dictionary)
```
