Q6. Explain the concept of generators in python and how they are defined.

```python
'''
Generator functions returns the solution one by one instead of alltogether.
normal functions generate the complete solution in one go,which takes time to process in case of uncountable series,
but a generator gives the solution one byb one so that an infinite series can be processed.
"yield" keyword is used in generator function which tells the code to return a solution one by one.
eg: Fibonacci sequence, is an infinite sequence which cannot be perocessed in one go. so we use a generator for this purpose.
'''
```

```python
#Fibonacci sequence:

def fib(n):
  a=0
  b=1
  for i in range(n):
    yield a
    a,b = b,a+b
```

```python
#calling the sequence.
f=fib(7)
```

Q10.What is the difference between 'map()' , 'reduce()'  and 'filter()' function in python?


```python
'''
Map():
--> Applies a given function to each item of an iterable(such as a list,tuple, or set)
       and returns a new iterable with the results.
    syntax:  map(function , iterable)
'''
#for eg:
numbers= [ 1,5,6,7]
square = list(map(lambda x : x**2 , numbers))
square
```

```python
'''
Filter():
--> constructs an iterator from elements of an iterable for which a function returns True.
    Syntax:  filter(function,iterable)
'''
#for eg
numbers= [ 1,5,6,7,10,14,31,16]
even_num = list(filter(lambda x : x%2==0 , numbers))
even_num
```

```python
'''
Reduce():
--> it applies a rolling computation to a sequential pairs of values in an iterable.
   Syntax:  reduce(function, iterable)
'''
#for eg:
numbers= [ 1,5,6,7,8,4,12,14,16,19,20]
sum_of_num = reduce(lambda x,y : x+y, numbers)
sum_of_num
```

Q4. Write a python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
  if n<=1:
    return False
  for i in range(2,n):
    if n%i==0:
      return False
  return True
  print(f"{n} is a prime number")
```

```python
print(is_prime(67))
print(is_prime(23))
is_prime(95)
```

```python
is_prime(9)
```
