2. How does broadcasting work in NumPy?

```python
'''
Broadcasting in NumPy allows us to perform arithmetic operations on arrays of different shapes without reshaping them.
It automatically adjusts the smaller array to match the larger array's shape by replicating its values
 along the necessary dimensions.
This makes element-wise operations more efficient by reducing memory usage and eliminating the need for loops.
'''
```

15.  How does NumPy handle multidimensional arrays ?

```python
'''
NumPy handles multidimensional arrays very efficiently, such as:

Homogeneous Data: NumPy arrays store elements of the same data type, which allows
                   optimized memory usage and faster computations.
Underlying Storage: Multidimensional arrays are stored in a contiguous block of memory,
                     similar to a one-dimensional array.
Shape and Strides: NumPy uses metadata (shape, strides, data type) to interpret
                     this contiguous block as a multidimensional structure.
           eg: A tuple indicating the size of each dimension (e.g., (3, 4) for a 3x4 matrix).
'''
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?

```python
import seaborn as sns

data = np.random.randn(100,5)
df= pd.DataFrame(data,columns= ['A','B','C','D','E'])
corr_matrix = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True , cmap= 'coolwarm', square=True)

plt.title('Correlation Matrix')

plt.show()
```

6. Create a DataFrame and add a new column based on an existing column.

```python
data= {'Name':['John','Anna','Peter','Linda'],
        'Age': [18,24,35,82],
        'Score':[85,90,78,92]}
df = pd.DataFrame(data)

df['Category']= pd.cut(df['Age'],bins=[0,25,35,100],labels=['Young','Adult','Senior'])

print("Original DataFrame : \n" , df[['Name','Age']])
print("\n DataFrame with the categorical column : \n" , df)
```

8. Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x,y1,label='Sine',color='blue')
plt.plot(x,y2,label='Cosine',color='red')
plt.plot(x,y3,label='Tangent',color='yellow')

plt.title('LINE PLOT WITH MULTIPLE LINES')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

9.  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
data = {
  'Name': ['John', 'Anna', 'Peter', 'Linda'],
 'Age': [18, 24, 35, 82],
 'Score': [85, 90, 78, 92]}
df= pd.DataFrame(data)

threshold = 85

filtered_df = df[df['Score']>threshold]

print('original dataframe :\n ',df)
print('\n \n Filtered dataframe :\n ', filtered_df)
```
