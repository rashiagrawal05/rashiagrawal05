Q1. What are the 5 key concepts of Object Oriented Programming(OOP)?


```python
'''
The 5 key concepts of object oriented programming in pyhton are :
1. CLASSES AND OBJECTS:
       In Python, a class is a template for crreating objects. A class defines the properties and behaviour of an object.
       An Object is an instance of a class, and has its own set of attributes(data) and methods(functions).

2. INHERITANCE:
      Inheritance is the mechanism by which one class can inherit the properties and behaviour of another class.
      The child class inherites all the attributes and methods of the parent class and can also add new attributes
      and methods or overrides the ones instead of inherited from the parent class.

3. ENCAPSULATION:
      Encapsulation is the concept of building data and methods that manipulates the data into a single unit, making it
      harder for other parts of the program to access or modify the data directly. In Python, Encapsulation is achieved
       using classes and objects.

4. ABSTRACTION:
      Abstraction is the concept of showing only the necessary information to the outside world while hiding the implementation
       details. In Python, abstraction can be achieved using abstract classes and interfaces.


5. POLYMORPHISM:
      polymorphism is the ability of an object to take on multiple forms. In Python, polymorphism can be achieved through
       method overloading or method overriding.

'''
```

Q2. Write a Python class for a 'Car' with attributes for 'make' , 'model' and 'year'. Include a method to display the car's information.

Q3. Explain the difference between instance methods and class methods . Provide an example of each.

```python
'''
Instance methods :
  this methods operate on an instance of the class and can access and modify instance attributes.
   this methods always take "self" as the first parameter.

Class methods :
  this methods operate on the class itself rather than any instance .
   this method takes "cls" as the first parameter and are decorated with @classmethods.
'''
```

```python
#for example :
class Animal:
  phyllum = "Chordata"
  def __init__(self,name,species,genera):
    self.name = name
    self.species = species
    self.genera = genera
  def describe(self):                              #this is an instance method.
    return f"{self.name} belongs to {self.genera} {self.species}"

  @classmethod
  def set_phyllum(cls,phylum_name):                #this is a class method.
    cls.phyllum = phylum_name
```

```python
dog = Animal("Dog","familiaris","Canis")

print(dog.name)
print(dog.species)

print(dog.phyllum)

print(dog.describe())         #calling instance method.

dog.set_phyllum("chordates")
print(dog.phyllum)            #calling class method.
```

Q5. What are the 3 types of access modifiers in python ? How are they denoted ?

```python
'''
1. Public :  Public access modifiers are accessible from anywhere in the program.
            They can be accessed using the dot notations. (eg- object.variable)
            No underscore prefix is used to denote public access modifiers.
            eg: (variable)
'''
```

```python
'''
2. Protected : Protected access modifiers are intended for internal use within the class or its subclasses.
              They can still be accessed from outside the class using the dot notation(eg- object._variable)
              the underscore indicates that the member is intended for internal use.
              A single underscore is used as a prefix to denote protected access modifiers.
              eg:(_variable)
'''
```

```python
'''
3. Private :  Private access modifiers are intended to be inaccessible from outside the class.
             They can be accessed outside only with the class name to ensure privacy of the variable(eg- _classname__variable)
              i.e. only programer can access the private variables
             A double underscore is used as a prefix to denote private access modifiers.
             eg: (__variable)
'''
```

Q6. Describe the five steps of inheritance in Python. Provide a simple example of multiple inheritance.

```python
'''
The five steps of inheritance in python are:
  1. Define a parent class or base class.
  2. Define a Child class that will inherit from the parent class.
  3. Use the inheritance syntax: the name of the parent class inside the parenthesis() to indicate inheritance .
  4. Override or extend Parent class members(methods or attributes) into the child class.
  5. Create an instance of the child class to use it as the members.
'''
```

```python
#example of multiple inheritance:

class Animal:                        #parent class 1
  def __init__(self,name):
    self.name = name
  def eat(self):
    print(f"{self.name} is eating.")

class Mammal:                       #parent class 2
  def __init__(self, hair_color):
    self.hair_color = hair_color
  def walk(self):
    print(f"the mammal with {self.hair_color} hair is walking.")

class Dog(Animal,Mammal):            # child class inherited from two parent classes.
  def __init__(self,name,hair_color,breed):
    Animal.__init__(self,name)
    Mammal.__init__(self,hair_color)
    self.breed = breed
  def bark(self):
    print(f"{self.name} the {self.breed} is barking. ")
```

```python
#creating instances
my_dog = Dog("Rex", "brown","Golden Retriever")
```

```python
my_dog.eat()
```

```python
my_dog.walk()
```

```python
my_dog.hair_color
```

```python
my_dog.bark()
```

Q7. What is the Method Resolution Order(MRO) in Python ? How can you retrieve it programmatically.

```python
'''
Method Resolution Order in python refers to the order in which a class's methods are searched when a method is invoked.
Generally used in case 0f multiple inheritance cases.
MRO is calculated on the basis of set of rules:
  1. Subclasses are searched before the parental class members.
  2. Parents are searched in the order they are listed in the class definition.
  3. If a parent class is listed multiple times, it is only searched once.
'''
```

```python
# MRO can be retrieved programatically using the  " mro() method "
# for eg:
class A:
  pass

class B:
  pass

class C(A):
  pass

class D(B,C):
  pass

print(D.mro())
```

Q8. Create an abstract base class 'shape' with an abstract method 'area()'. Then create two subclasses 'ractangle' and 'circle' that implements the 'area()' method .

```python
import abc
class Shape:
  @abc.abstractmethod
  def area(self):
    pass

class Rectangle(Shape):
  def area(self,length,bredth):
    return f"the area the the rectangle is {length * bredth} sq.units"

class Circle(Shape):
  def area(self,radius):
    return f"the area of the circle of radius {radius} units is {(22/7 * radius**2)} sq.units"
```

```python
rect = Rectangle()
c = Circle()

print(rect.area(20,12))
print( c.area(70))
```

Q10. Implement encapsulation in a 'Bank Account' class with private attributes for 'balance' and 'account number'. Include methods for deposit, withdrawal, and balance inquiry.

```python
class Bank_Account:
  def __init__(self,balance,account_num):
    self.__balance = balance
    self.__account_num = account_num

  def deposit(self,amount):
    self.__balance = self.__balance + amount
    print("the amount is deposited")

  def withdraw(self,amount):
    if (amount < self.__balance):
      self.__balance= self.__balance-amount
      print(f"₹{amount} has been deduced from your account")
    else:
      print("insufficient balance")
    return self.__balance

  def balance_inquiry(self):
    return f"your current balance for the account number {self.__account_num} is ₹{self.__balance} "
```

```python
acc = Bank_Account(10000, 64982321132)
print(acc.balance_inquiry())

acc.deposit(1500)
print(acc.balance_inquiry())
acc.withdraw(7500)
print(acc.balance_inquiry())
acc.withdraw(4000)
print(acc.balance_inquiry())
```

Q11. Write a class that overrides the '_ _str_ _' and '_ _add_ _ ' magic methods. What will these methods allow you to do?

```python
class Vector:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __str__(self):
    return f"Vector({self.x},{self.y})"

  def __add__(self,other):
    if isinstance(other,Vector):
      return Vector(self.x + other.x , self.y + other.y)
    else:
      raise TypeError("Unsupported operand type for '+' ")
```

```python
v1 = Vector(2,4)
v2 = Vector(4,5)
```

```python
print(v1)
```

```python
print(v2)
```

```python
v3 = v1 + v2
print(v3)
```

Q12. Create a decorator that measures and prints the execution time of a function.

```python
import time

def timer_decorator(func):
  def timer():
    start = time.time()
    func()
    end = time.time()
    print("the time taken to execute this function is ", end-start)
  return timer
```

```python
@timer_decorator
def func_test():
  print(100/4)
```

```python
func_test()
```

Q13. Explain the concept of the Diamond in multiple inheritance. How does python resolve it?

```python
'''
In multiple inheritance, the diamond problem refers to a situation where two classes inherit from a common base class,
and then another class inherits from both of those classes.
this appears a diamond-shaped inheritance graph.
  A                 Here, A is the base class.
 / \                 B and C inherit from A (i.e. Hierarchial inheritance)
B   C               class D inherites from both B and C (i.e. Multiple inheritance)
 \ /
  D

The problem arises when B and C overrides a method from A, and class D tries to call the method.
  then problem occurs that which implementation of method should D use i.e. from B or from C.

To resolve this problem, Python uses "Method Resolution Order (MRO)"
 which determines the order in which to search for methods in the inheritance graph.

For above problem, the MRO would be
  D > B > C > A > object

'''
```

Q14. Write a class method that keeps track of the number of instances created from a class.

Q15. Implement a static method in a class that checks if a given year is a leap year.

```python
class Year:
  def __init__(self):
    pass
  @staticmethod
  def is_leap(year):
    return (year % 4 == 0 ) and (year % 100 != 0 or year % 400 == 0)
```

```python
print(Year.is_leap(2000))
print(Year.is_leap(1987))
print(Year.is_leap(2000))
print(Year.is_leap(5000))
```
